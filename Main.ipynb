```python
import random
import math

# Random bit generator based on the number_of_bits


def randomBits(number_of_bits):
    r = ''
    for i in range(number_of_bits):
        r += random.choice('01')
    return r


# Representation of individual
number_of_bits = 10

# Hyper parameter
population_size = 20
individuals_in_selection = 5
number_of_generations = 10
crossover_probability = 0.9  # 90%
mutation_probability = 1/number_of_bits


# Initiate the population
population = []

for individual in range(population_size):
    population.append(randomBits(number_of_bits))


def fitness(individual):
    f = 0
    for i in range(len(individual)):
        if individual[i] == '1':
            f += 1
    return f


def one_point_crossover(individual_one, individual_two):
    cut = random.randint(1, number_of_bits-1)
    offspring = individual_one[0:cut] + individual_two[cut:]
    return offspring


def bit_flip_mutation(individual):
    mutation = ''

    for i in range(len(individual)):
        if random.random() < mutation_probability:
            if individual[i] == '0':
                mutation += '1'
            else:
                mutation += '0'
        else:
            mutation += individual[i]

    return mutation

# My tournament_selection
def tournament_selection(population):

    max_fitness_of_individual = -math.inf  # Initialising maximum fitness

    # From population we take the amout individuals_in_selection of random individuals.
    for i in range(individuals_in_selection):
        random_individuals_index = random.randrange(len(population))

        # Then individual with highest fitness is selected for crossover
        if fitness(population[random_individuals_index]) > max_fitness_of_individual:
            max_fitness_of_individual = fitness(
                population[random_individuals_index])
            individual_with_max_fitness = random_individuals_index

    return individual_with_max_fitness  # Inidivual with highest fitness return

# Iterate the generations. For each generation chance the population based on selection and what not
```
